```python
import numpy as np
import time
import pandas as pd
from pandas.plotting import register_matplotlib_converters
import matplotlib.pyplot as plt
import ssl


register_matplotlib_converters()

ssl._create_default_https_context = ssl._create_unverified_context
```

```python
ticker = input()
now = str(round(time.time()))
history_start = str(round(time.time()-5*365*24*60*60))
url = "https://query1.finance.yahoo.com/v7/finance/download/"+ticker+"?period1="+history_start+"&period2="+now+"&interval=1d&events=history"
url
```

```python
df = pd.read_csv(url)
df
```

```python
df["Date"] = pd.to_datetime(df["Date"])
data = df.filter(['Date','Close'])
data.set_index("Date", inplace=True)
data['SMA200'] = data.rolling(window=200).mean()
data
```

```python
fig = plt.figure(figsize=(18, 10))
plt.title(ticker)
plt.xlabel("Datum")
plt.ylabel("Preis")
plt.plot(data.index, data["Close"], label="Schlusskurs")
plt.plot(data.index, data["SMA200"], label="SMA200")
plt.legend(loc='upper left')
plt.show()
```
