```python
import numpy as np
import matplotlib.pyplot as plt

plt.plot([-1, -4.5, 16, 23, 15, 59])
plt.show()
```

```python
li = np.linspace(0, 10, 100)
plt.plot(li)
plt.show()
```

```python
si = np.sin(li)
plt.plot(si)
plt.show()
```

```python
plt.figure(figsize=(15.0, 10.0))
plt.xlabel('number')
plt.ylabel('sin(number)')
plt.plot(si, ">b")
plt.plot(li, color='green', marker='o', linestyle='dashed', linewidth=1, markersize=1)
plt.show()
```

```python
data = {'a': np.arange(20),
        'c': np.random.randint(0, 50, 20),
        'd': np.random.randn(20)}
data['b'] = data['a'] + 10 * np.random.randn(20)
data['d'] = np.abs(data['d']) * 100
data
```

```python
plt.figure(figsize=(15.0, 10.0))
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```
