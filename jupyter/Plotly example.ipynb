```python
import pandas as pd
import plotly.express as px

data_frame = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2014_apple_stock.csv', parse_dates=True)
data_frame.set_index('AAPL_x', inplace=True)
data_frame.head()
#data_frame['AAPL_y']
```

```python
fig = px.line(data_frame, x = data_frame.index, y = 'AAPL_y', title='Apple Share Prices over time (2014)')
fig.show()
```

```python
#data_frame.set_index('AAPL_x', inplace=True)
#data_frame.head()
```

```python
data_frame['rolling20'] = data_frame.rolling(window = 20).mean()
data_frame.head()
```

```python
fig = px.line(data_frame, x = data_frame.index, y = 'rolling20', title='Apple Share Prices over time (mean 20d)')
fig.show()
```
